### Multi Line plot

```python
a=np.arange(0, 6, 0.1)
b=np.arange(1, 8, 0.1)
c=np.arange(2, 10, 0.1)
plt.plot(a, [a*2 for a in a])
plt.plot(b, [b*3 for b in b])
plt.plot(c, [c*4 for c in c])
```

### Adding Grid and Axis

There are two ways to pass parameters to axis(): by a list of four values or by keyword arguments.

1. four values of keyword arguments [xmin, xmax, ymin, ymax]
2. specific keyword arguments, for example: plt.axis(xmin=NNN, ymax=NNN)

```python
x=np.arange(0, 6, 0.1)
y= np.arange(0, 8, 0.1)
z=np.arange(0, 10, 0.1)
plt.plot(x, [x*2 for x in x])
plt.plot(y, [y*3 for y in y])
plt.plot(z, [z*4 for z in z])
plt.grid(True)
print(plt.axis())        # It shows the value of x axis and y axis (Minimum and Maximum values)
```

```python
x=np.arange(0, 6, 0.1)
y= np.arange(1, 8, 0.1)
z=np.arange(2, 10, 0.1)
plt.plot(x, [x*2 for x in x])
plt.plot(y, [y*3 for y in y])
plt.plot(z, [z*4 for z in z])
plt.grid(True)
#plt.axis([-0.2,12,-1,50])     # It is for setting axis
plt.axis(xmin=-0.2, ymax=50)   
```

### Adding Legend

A legend is an area describing the elements of the graph.

In the matplotlib library, there's a function called legend() which is used to place a legend on the axes.

```python
days=[1,2,3,4,5,6,7]
max_temp=[50,51,52,48,47,49,46]
min_temp=[43,42,40,44,33,35,37]
avg_temp=[45,48,48,46,40,42,41]

plt.plot(days, max_temp, label="max")
plt.plot(days, min_temp, label="min")
plt.plot(days, avg_temp, label="average")
plt.grid(True)

# plt.legend()          # Adding legend with by Default position which is Best
# plt.legend(loc=6)     # Adding legend at different positions
plt.legend(loc="best",shadow=True,fontsize="large",facecolor="lightblue",edgecolor="black") # For making legend box look more attractive
```

### Control Line Styles
`-` solid line<br>

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([4, 10, 7, 14])
y3=  np.array([2, 7, 5, 11])

plt.plot(y1,'b')                
plt.plot(y2 ,'m-.')
plt.plot(y3, 'r--')              # Adding colours and line style in the lines of the graph

plt.grid(True)
```

## Bar Chart

With Pyplot, we can use the bar() function to draw bar graphs

```python
d1 = {'C':20, 'HTML':15, 'Java':30,'Python':35}
courses = list(d1.keys())
val = list(d1.values())
  
fig = plt.figure(figsize = (6, 4))                              # Here fig size parameters are width and hieght in inches
 
# creating the bar plot
plt.bar(courses, val, color =['orange','green'], width = 0.7)   # Here width is used to set size of bars
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
d1 = {'C':20, 'HTML':15, 'Java':30,'Python':35}
courses = list(d1.keys())
values = list(d1.values())
  
fig = plt.figure(figsize = (6, 4))        # Here fig size parameters are width and hieght in inches
 
# creating the bar plot
plt.barh(courses, values, color='purple',height=0.8)  # here height is used for setting the thickness of bars in horizontal orientation
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

## Pie Charts

With Pyplot, you can use the pie() function to draw pie charts

```python
a = [35, 25, 25, 15]
mylabels = ["Apples", "Mangoes", "Bananas", "Cherries"]

plt.pie(a, labels = mylabels)
plt.show() 
```

Some important parameters

`explode`: Specify the offset radius fraction the wedge has from the center of the pie<br>
`colors`: This is a list of Matplotlib colors, cyclically used to color the wedges.<br>
`labels`, `labeldistance`: This is a list of labels<br>
`autopct` : Label wedges with their numeric values.<br>
`pctdistance`: pctdstance is the ratio between the center of the pie and the start of the text.<br>
`shadow`: This draws a shadow for wedges or pie.

```python
a = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Grapes", "Orange"]
myexplode = [0.2, 0.04, 0.1, 0]
mycolors = ["red", "yellow", "#90EE90", "#FFA500"]


plt.pie(a, labels = mylabels, explode = myexplode,autopct='%1.1f%%',shadow=True,colors=mycolors)
plt.show() 
```

## Subplots

With the subplots() function we can draw multiple plots in one figure

The subplots() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the particular plot.

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```
